##### 착수보고서 
- 착수보고서란 개발 프로세스에서 클라이언트가 무엇을 요청했는데, 언제까지 가능하고, 어떤 기능을 쓸거고 작성하는 일종의 계약서와 같음
- 제안서에 있는것 중에 어떤게 가능하다. 어떤게 필요하다 어떤건 불가능하다 작성해줘야한다.
- 안되면 왜 안되는지 다른걸로 바꿀 방법이 있으면 그 방법을 제안해준다.
- 분량을 굳이 늘릴려고 하지 말고 핵심과 요지만 정확히 집어낸다.
- 요약 문단을 꼭 적어주자. 두괄식으로. 말하고자 하는 주제를 정확하게 (읽는 사람의 관심이 사라지기 전에 최대한 빨리 말해줘야함)
- 요구사항을 나열하고 가능 / 불가능 (해결방안)을 작성해놓으면 거의 완성된다.
- 기술현황 / 관련기술을 제시하면 기술 문서가 들어가야한다. 이를 별첨에 넣는다.
- 본 업무를 위해서 어떤걸 할 것인가, 무엇을 할 것인가, 어떻게 할 것인가를 상세하게 설명해야한다.
- 요구사항을 모두 충족해야한다. (충족하지 못하면 협상해야함)
- 진행 중 하지 못하는 부분이 생긴다면 바로 담당자와 협의 후 변경 혹은 수정해야한다. 또한 이 내용을 착수보고서에 우선 수정을 하고 진행해야한다.

- 역할 분배(관리 책임자), WBS(word breakdown structure)는 최적으로 짜야함

- 가설 : ~~조건을 만족하면 ~한 사람들은 ~할 것이다 와 같이 생각하거나, 실제로 존재하는 통계나 지식, 데이터를 사용함
- 어떤 기법이 사용되고 어떤 절차로 수행되는지 workflow를 첨부해야함. 또한 동떨어진 프로그램(배우지 않았거나)을 사용한 경우 정보를 적어줘야하고, 사용한 라이브러리나, 패키지가 있다면 라이센스를 작성해줘야함
- 결과물 / 답을 작성할 땐 객관적인 사실만 제시해야함.(숫자로 표현할 수 있는 것)
- 가설 검증 결과, 정량적 데이터 제시(그래프)
- 인용(외부문서, 파일, 뉴스, 기관, 패키지, 파이썬, 등)

##### 데이터 전처리 (Data preprocessing)

- 데이터 : 정제되지 않은 사실 
- 정보 : 의미가 부여되어있고 조직화된 데이터
- 즉 데이터에 의미와 조직화를 해주면 정보가 딘다.
- 데이터는 분석에 쓸 수 없고 정보는 쓸 수 있기 때문에 전처리 해줘야한다.
- 예전에는 데이터 수집이 큰 매리트가 있었지만 지금은 데이터가 많이 풀려있기 때문에 개인이나 소규모 단체에서는 정제가 더 중요하다.

- 데이터를 정보로 바꾸기 위해서 데이터 전처리를 한다.
- 데이터 전처리는 데이터의 문제를 해결하려고 한다.
1. 데이터 누락 : NaN (단, 누락된 자료가 엄청나게 적어서 무시해도 되거나, 누락된 데이터를 제외하고 사용해도될 때 처리 안해도된다.)
2. 데이터 오차 : A씨 할아버지 나이 395세
3. 형태를 벗어난 값 : 올해 연도 '4월2일'
- 위와 같은 예시처럼 데이터에 문제가 있어서 분석을 하기 어려울 때 데이터 전처리를 하면 된다.

##### 데이터 전처리 종류
1. 데이터 클리닝
2. 데이터 통합
3. 데이터 변환
4. 데이터 축소
5. 데이터 이산화
6. 결측치 및 이상치 처리
7. 일관성 있는 데이터로 변환(표준화, 정규화)

##### 데이터 클리닝의 종류
1. 중복 데이터 삭제(가장 먼저적힌거 남김, 가장 최신기록 남김)
2. 관련 없는 데이터 삭제(지역별 온도 데이터 활용하는데, 예산이 들어가있다던지)
3. 구조적 문제 수정(주민번호 안에 생년월일, 성별이 들어있는데 또 입력하라고 한다던지)
4. 대소문자 일치(영어에서 주로 발생함)
5. NA표기 통일 (fillna로 처리함)
6. 이상치(outlier) 처리
7. 누락(NA)데이터 처리
8. 수정된 데이터 평가

##### 데이터 통합
- 출처가 다양한 데이터를 단일화한다.
1. 동일한 데이터를 통합 (다양한 데이터셋에 같은 데이터가 포함되어있다면 같은 데이터 기준으로 통합)
2. 두 데이터 간의 차이를 확인
3. 같은 의미를 가지는 데이터 검토
4. 같은 데이터를 포함한 정보 처리

##### 데이터 변환
- 분석이 불가능하거나 어려운 형태를 분석 가능한 형태로 변환
1. 단위 변환 (화폐 등)
2. 문자열의 경우엔 파싱
3. 형태 변환
4. 여러 의미를 가진 데이터의 경우 분리 저장

##### 데이터 축소
- 데이터를 저차원, 저용량으로 축소
1. 차원축소 (PCA t-SNE 등) : 원래 데이터의 큰 틀은 유지되지만,, 손실이 있다.
2. 유효숫자 변경
3. 수량 감소
4. 데이터 속성 제거
5. 클러스터링, 샘플링
6. 데이터 용량 축소

##### 데이터 이산화 
- 연속형 데이터를 범주형 데이터로 변경
1. 기계학습에 많이 사용된다.
2. 차이가 큰 값에 가중치 부여, 차이가 적은 값의 연관관계 부여를 예방한다.
3. 컴퓨터가 판단할 수 없다는 특성에 근거

##### 결측치 이상치 처리
- 분석 불가능하거나 크게 영향을 주는 데이터 처리
1. 결측치 처리 (제로필, 평균값 또는 중간값으로 대체)
2. 이상치 처리 (제거, 너무 많이 지워야한다면 대체) -> 대체는 보통 컷오프로 한다. (분포쪽으로 들고온다. 내용이 왜곡되는 현상을 줄인다. 하지만 너무 많은 데이터가 옮겨진다면 데이터의 신뢰도가 떨어진다.)
- 이외에도 기계학습을 통한 대체를 이용하기도 한다.

##### 표준화(Standard-), 정규화(Norm-)
- 데이터의 분포에 따른 특성 제거 또는 통계를 위한 변환
1. 표준화
- 정규분포의 속성을 갖도록 데이터 분포가 변화된다.
- 데이터 범위의 제한이 없다.
- Z-score
2. 정규화
- 데이터 분포는 유지된다.
- 데이터 범위의 제한이 생긴다.
- min-max scaling
- 연산량을 줄이기 위해서, 기계학습 모델을 사용하기 위해서 사용된다.

```python
import pandas as pd
```

```python
media = pd.read_csv('../시청자미디어재단_체험형미디어교육_20240105.csv',encoding='cp949')
import matplotlib.pyplot as plt
import matplotlib
plt.rc('font', family='Malgun Gothic') #윈도우에서 한글 사용할 수 있게 해줌
matplotlib.rcParams['axes.unicode_minus'] = False #마이너스 기호 오류 해결하기 (유니코드의 -기호를 사용하지 않겠다.)

print(media.head(5))
```

##### 결측치 제거하기

```python
print(media['센터명']) #결측치 관측되는중 NaN
print(media['센터명'].unique()) #어떤 데이터가 있는지 확인
```

```python
#결측치를 제거하자 !
#media.fillna(0) 하면 NaN값이 0으로 된다.
media = media.dropna(how='all',axis=0) #how로 조건을 전체로 설정해서 모두 NaN값일때, axis = 0 이면 row기준으로 삭제한다.
print(media) #지워진 것을 확인할 수 있다.
```

##### 중복되는 값 슬라이싱으로 제거하기

```python
#시청자 미디어 센터가 중복되니까 센터명에서 시청자미디어센터를 지워보자 !
media['센터명'] = media['센터명'].apply(lambda x : x[:2])
print(media) #이렇게 지울 수 있다. 
```

##### dictionary로 값 바꾸기

```python
#유아 -> 미취학
#5~7세 미취학
#일반시민 일반시민
#초등생이상 일반시민
#중등 중등
dict1 = {'유아':'미취학','5~7세':'미취학','일반시민':'일반시민','초등생 이상':'일반시민','중등':'중등'}
media['교육대상구분'] = [dict1[i] for i in media['교육대상구분']]
print(media['교육대상구분'].unique())
```

```python
#위의 코드와 같이 데이터를 변환할 수 있다.
def change(x):
    if x == '월,화,수,목,금,토':
        return '주중'
    return x
media['운영요일'] = media['운영요일'].apply(lambda x : change(x))
print(media)
```

```python
print(media['운영요일'].unique())
```

##### 날짜처리

```python
from datetime import datetime #파이썬 날짜처리 가능 !
#str로 되어있는 날짜를 날짜 형식에 맞춰보자
media['교육시작일자'] = media['교육시작일자'].apply(lambda x : datetime.strptime(x,'%Y-%m-%d'))
```

```python
print(media['교육시작일자'])
```

```python
media['교육종료일자'] = media['교육종료일자'].apply(lambda x : datetime.strptime(x,'%Y-%m-%d'))
print(media['교육종료일자'])
```

```python
media['교육진행기간'] = media['교육종료일자']-media['교육시작일자'] #몇일동안 진행했는지 알 수 있다.
media['교육진행기간'] = media['교육진행기간'].apply(lambda x : x.days+1)
# days를 해줘야지 time stamp가 숫자로 된다.
print(media['교육진행기간'])
```

##### 공휴일 관련 정보를 알고 싶다면, holiyday 패키지로 해결가능하다.
