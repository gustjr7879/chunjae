### 데이터 마이닝(Data mining)
축적된 정보 중 가치있는 정보를 발굴하는 과정     
단순 데이터 수집 뿐만 아니라 가공 및 분석과정을 일컫는다.    

#### 데이터 마이닝의 절차
- 문제 정의 : 목표, 요구사항(무슨 데이터는 필요하고,, 무슨 데이터는 필요없고,,), 구현계획     
- 데이터 수집 및 준비 : 매개변수의 선택, 데이터 클리닝 작업    
- 모델 구축 및 평가 : 분석 기법 선택 및 데이터 보정
- 모델 배포 : 새로운 데이터에 모델 적용, 모델 정보 추출 등     

#### 웹 크롤링과 웹 스크래핑
웹 크롤링(Web Crawling)과 웹 스크래핑(Web Scraping)은 구분되는 개념이지만 실무에서는 구분 없이 사용한다.    
- 웹 크롤링 : 웹 크롤러가 다양한 페이지를 인덱싱한다.    
- 웹 스크래핑 : 웹사이트의 데이터를 추출해서 저장한다.      

#### 웹 크롤링
웹 구조에 따른 파싱(parsing) 작업이 이루어지므로 단순 복사-붙여넣기와 다른 특성을 보인다.

위와 같이 엄청나게 간단하게 크롤링해올 수 있다.     
물론 세세한 크롤링 설정을 할 수 없다는 제약이 있지만 간단하게 제목, 내용, 등을 파악하기엔 가장 편하다

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

#알라딘 url
page = 200000 #순위별로 페이지 다름
url = f'https://aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page={page}&cnt=1000&SortOrder=1'

req = requests.get(url) #html파일 요청해주는 역할 : request
content = req.content
print(content) #content = html 파일 내용
soup = BeautifulSoup(content,'html.parser')
```

```python
#html 파일 분석 결과 a.bo3 클래스의 b태그에 책 이름이 들어있다.
#또한 span.ss_p2 의 span에 가격이 들어있다.
book_names = soup.select('a.bo3 b') #a.bo3클래스 b태그
prices = soup.select('span.ss_p2 span') #가격
print(book_names[0].text)
print(int(prices[0].text.replace(',','')))
```

```python
# 크롤링해서 제목과 가격만 dataframe으로 저장
df = pd.DataFrame(columns=['책제목','가격'])
all_name = []
all_price = []
for i in range(1,4):
    page = i
    url = f'https://aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page={page}&cnt=1000&SortOrder=1'
    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content,'html.parser')
    book_names = soup.select('a.bo3 b') #a.bo3클래스 b태그
    prices = soup.select('span.ss_p2 span') #가격
    length = len(book_names)
    for j in range(length):

        all_name.append(book_names[j].text)
        all_price.append(int(prices[j].text.replace(',','')))
leng = len(all_name)
df['책제목'] = [all_name[j] for j in range(leng)]
df['가격'] = [all_price[j] for j in range(leng)]

print(df)
df.to_csv('./알라딘크롤링결과.csv')
```

### 동적 페이지에서 크롤링하기
알라딘 중고서적과 같이 하나의 페이지 당 하나의 주소로 구성되어있고 변하지 않는 정적인 페이지는 Beautifulsoup을 사용하면 간단하게 크롤링할 수 있었다.      
하지만 최근 페이지들은 대부분 반응형, 즉 사용자와 상호작용하면서 구성되어있는 동적인 페이지가 많다.     
이런 동적인 페이지에서 크롤링을 하기 위해서는 **Selenium**을 사용한다.     
이번에는 네이버 지도에서 크롤링하는 것과 네이버 뉴스 기사 댓글(더보기를 통해서 계속 불러올 수 있는)을 크롤링해보도록 하자

```python
import pandas as pd
from selenium import webdriver #크롬창을 조종하기 위함
from selenium.webdriver.common.by import By #웹사이트의 구성요소를 선택하기 위함
from selenium.webdriver.support.ui import WebDriverWait #웹페이지가 전부 로드될때까지 기다리기 위함
from webdriver_manager.chrome import ChromeDriverManager #크롬에서 크롤링을 하기 위해서 웹드라이버를 설치하는 모듈
from selenium.webdriver.support import expected_conditions as EC #크롬에서 어떤 부분의 상태를 확인하는 모듈
import time
```

```python
driver = webdriver.Chrome() #웹 드라이버를 설치하고 조종할 수 있는 크롬창 오픈
```

```python
#네이버 지도에서 원하는 키워드를 검색하자
url = 'https://map.naver.com/v5/search/서울 노원구 카페'
driver.get(url) #나는 서울 노원구 카페를 검색어로 사용할 것이다.
```

- 위 코드를 실행하면 열려있던 조종되는 크롬창에서 서울 노원구 카페를 검색한 것을 확인할 수 있다.
- 즉 get이 url을 받아서 검색을 해준다고 보면 된다. 이 기능은 다른 홈페이지로 이동할때도 요긴하게 사용된다.
- 우리가 웹페이지에 접속할 때 인터넷 회선, 서버 문제로 접속시간이 소요된다.
- 이런 문제를 고려하지 않고 코드를 바로 실행시켜버리면 문제가 발생한다.
- 문제를 해결하기 위해서 로딩이 다 끝날때까지 기다리는 나름 무식한 방법을 사용한다 ㅎ
- Implicity wait와 Explicity를 사용해서 로딩을 기다리는데, 전자는 정해진 시간만큼 기다리는 방법이고, 시간을 넉넉하게 설정하면 에러가 발생하지 않아서 사람이 신경쓸 일이 별로 없다.
- 후자의 방법은 인터넷 상태가 변동이 클때나 특정한 구성요소가 로딩되는 즉시 코드를 실행할 때 사용된다.
- 지도의 경우 검색창만 다 로딩되면 되니까 Explicity 방법을 사용할 수 있다.

```python
try:
    element = WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.CLASS_NAME,"input_search"))
    ) #네이버 지도의 검색창은 'input_search'라는 클래스 이름으로 설정되어있다.
finally:
    pass
```
