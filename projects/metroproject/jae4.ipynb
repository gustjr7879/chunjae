```python
import pandas as pd

confuse_data = pd.read_csv('../서울교통공사_지하철혼잡도정보_20221231.csv',encoding='cp949')
print(confuse_data["요일구분"].unique())
#
```

##### 처리할 것 
1. 시간대 별 혼잡도 순위 분석
2. 요일 구분을 평일을 별로 따로따로 만들기
3. 역별 구를 추가해야함
4. 시간대 별 구 별 혼잡도 분석 -> 사람들이 어디로 출근/퇴근하는지 파악

```python
print(confuse_data["출발역"].unique())
```

```python
gu_data = pd.read_csv('../서울교통공사_자치구별지하철역정보_20231031.CSV',encoding='cp949') #파일을 불러옴 (어떤 지하철역이 무슨구에 해당하는지 알기 위해서)
station_list = gu_data['해당역(호선)'] 
station_gu = gu_data['자치구']
fix_station_list = []
#print(station_list)
import re
for i in station_list:
    i = re.sub('[(\d)]','',i) #파일에 마들역(7)과 같이 호선까지 표기되어있었기 때문에 처리해줌
    i = list(map(str,i.split(', '))) #하나의 리스트로 묶여있던 것을 나눠서 한 리스트에 여러 요소로 저장
    fix_station_list.append(i)
```

```python
print(fix_station_list)
print(station_gu)
for i in range(len(fix_station_list)):
    for j in fix_station_list[i]:
        
        confuse_data.loc[confuse_data['출발역']==j,'자치구'] = station_gu[i]
```

```python
confuse_data.loc[confuse_data['출발역'] == '미아삼거리','자치구'] = '강북구'
confuse_data.loc[confuse_data['출발역'] == '별내별가람','자치구'] = '남양주'
confuse_data.loc[confuse_data['출발역'] == '서울역','자치구'] = '용산구'
confuse_data.loc[confuse_data['출발역'] == '신천역','자치구'] = '송파구'
confuse_data.loc[confuse_data['출발역'] == '신촌(지하)','자치구'] = '마포구'
confuse_data.loc[confuse_data['출발역'] == '오남','자치구'] = '남양주'
confuse_data.loc[confuse_data['출발역'] == '올림픽공원(한국체대)','자치구'] = '송파구'
confuse_data.loc[confuse_data['출발역'] == '을지로3가','자치구'] = '중구'
confuse_data.loc[confuse_data['출발역'] == '을지로4가','자치구'] = '중구'
confuse_data.loc[confuse_data['출발역'] == '종로3가','자치구'] = '종로구'
confuse_data.loc[confuse_data['출발역'] == '종로5가','자치구'] = '종로구'
confuse_data.loc[confuse_data['출발역'] == '진접','자치구'] = '남양주'
confuse_data.loc[confuse_data['출발역'] == '총신대입구','자치구'] = '동작구'

#누락된 부분들 추가해주기
```

```python
dropconfuse = confuse_data.drop(['역번호','상하구분','연번'],axis=1)
# 요일 구분하기
pyung = dropconfuse['요일구분'] == '평일'
pyungil = dropconfuse[pyung].reset_index()
print(pyungil)
ju = dropconfuse['요일구분'] != '평일'

jumal = dropconfuse[ju].reset_index()
print(jumal)
```

```python
pyungil = pyungil.drop(['요일구분'],axis=1)
pyungil_gu = pyungil['자치구']
pyungil = pyungil.drop(['자치구'],axis=1)
print(pyungil_gu)
print(pyungil)
grouped_df = pyungil.groupby('출발역').aggregate('mean') #일단 상행선 하행선은 무의미해서 groupby로 묶어주기 (혼잡도평균)
```

```python
grouped_df = grouped_df.drop(['index','호선'],axis=1)
print(grouped_df)
#grouped_df : 출발 역 시간별 혼잡도
```

```python
import plotly.express as px
#출 퇴근시간에만 관심이 있어서 나머지 시간대 제외
etc_df = grouped_df.drop(['5시30분','6시00분','6시30분', '10시30분','11시00분','11시30분','12시00분','12시30분','13시00분','13시30분','14시00분','14시30분','15시00분','15시30분','16시00분','16시30분','20시30분','21시00분','21시30분','22시00분','22시30분','23시00분','23시30분','00시00분','00시30분'],axis=1)
#출근시간 퇴근시간으로 색깔 나눠서 확인하기 편하게함
color_discrete_map = {'7시00분': 'blue', '7시30분': 'blue', '8시00분': 'blue', '8시30분': 'blue', '9시00분': 'blue', '9시30분': 'blue', '10시00분': 'blue',
                       '17시00분' :'red', '17시30분' :'red', '18시00분' :'red', '18시30분' :'red', '19시00분' :'red', '19시30분' :'red', '20시00분' :'red'}
fig = px.scatter(etc_df[:20],color_discrete_map=color_discrete_map) #지하철 역 별 시간대에 따른 혼잡도 표시
fig.show()
```

##### 30개 역 출퇴근 시간 혼잡도 확인
1. 퇴근시간이 제일 높은 역은 일자리가 많은 곳일 가능성이 높다.(18-19시)
2. 출근시간이 제일 높은 역은 주거지가 많을 곳일 가능성이 높다.(9-10시)
3. 애매한 시간에 높은 역은 환승역일 가능성이 높다(8시, 18시30분)
4. 전체적으로 낮은 경우에 종점, 혹은 출발역 주변일 가능성이 높다

```python
for i in range(len(fix_station_list)):
    for j in fix_station_list[i]:
        
        etc_df.loc[etc_df.index==j,'자치구'] = station_gu[i] #구 다시 추가해주기
```

```python
#구 별로 출근시간대 퇴근시간대 자료 만들어주기
ff = etc_df.copy()
ff.index = [ff['자치구']] #index를 바꿔주는 방법이다.
ff = ff.drop(['자치구'],axis=1)
test_df = ff.groupby('자치구').aggregate('mean') 
print(test_df)
fig = px.scatter(test_df,color_discrete_map=color_discrete_map)
fig.show()
```

##### 30개 역 출퇴근 시간 혼잡도 확인 (자치구별)
1. 퇴근시간에 주로 높은 역은 일자리가 많은 구일 가능성이 높다.
2. 출근시간에 주로 높은 역은 주거지가 많은 구일 가능성이 높다.(9-10시)
3. 사람들은 보통 8시에 지하철을 탑승해서 6시에 지하철을 탑승한다.
4. 거리가 있는 의정부시와 하남시에서는 서울에 있는 구 보다 30분 일찍 탑승한다.
5. 일자리가 강남구, 고양시, 금천구, 서초구, 종로구, 중구에 몰려있다
6. 강북구 광명시 구로구, 동작구, 서대문구, 성북구, 도봉구, 양천구 와 같은 구들은 대부분 주거단지 형태를 띄는 것을 알 수 있다.
