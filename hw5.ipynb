```python
import requests
from bs4 import BeautifulSoup

data = requests.get('https://gustjr7879.github.io/hyunskki/chunjae8/') #내 블로그로 설정

soup = BeautifulSoup(data.content, 'html.parser')

print(soup) #크롤링 테스트 완료
```

```python

names = soup.select('h1') #크롤링 완벽하게 되는지 체크
print(names)
```

```python
import pandas as pd
df = pd.DataFrame() #저장할 데이터프레임 설정 

all_name = []
all_date = []
all_contents = []
all_link = []
for i in range(1,9): #1페이지부터 8페이지까지 천재교육 강의내용 정리 블로그 탐색
    url = f'https://gustjr7879.github.io/hyunskki/chunjae{i}/' #formatting으로 탐색
    req = requests.get(url) #request
    soup = BeautifulSoup(req.content, 'html.parser') 
    names = soup.select('h1') #제목 가져옴
    date = soup.select('h4')[0] #날짜 가져옴
    contents = soup.select('#markdown-toc > li > ul')[0] #같은이름 태그가 너무 많아서 상세하게 설정
    print(contents) #잘 가져와지나 확인
    link = soup.select('body > div.wrapper > div.article > p:nth-child(3) > a') #위와 동일하게 태그가 너무 많아서 상세하게 설정 
    all_name.append(names[0].text) #h1에서 첫번째만 사용
    all_date.append(date.text)
    all_link.append(link)
    all_contents.append(contents.text)
    
```

```python
df = pd.DataFrame({'제목' : all_name, '작성날짜': all_date,'링크':all_link,'학습 콘텐츠':all_contents}) #데이터프레임 설정

df.to_csv('./test.csv',index=False) #csv파일로 저장
```

```python
url = 'https://gustjr7879.github.io/hyunskki/blog/' #전체 블로그에서 진행
req = requests.get(url)
soup2 = BeautifulSoup(req.content)
all_contents = soup2.select('a')[5:-1] #필요없는 링크 제거
print(len(all_contents))
```

```python
all_name = []
all_date = []
all_content = []
all_link = []
print(type(all_contents[0])) #전체 블로그 글 몇개인지 파악
for i in range(len(all_contents)):
    url = all_contents[i]['href'] #tag에서 url만 가지고옴
    req = requests.get(url) #request
    soup = BeautifulSoup(req.content, 'html.parser') 
    names = soup.select('h1') #제목 가져옴
    date = soup.select('h4')[0] #날짜 가져옴
    contents = soup.select('#markdown-toc > li > ul > li > a') #같은이름 태그가 너무 많아서 상세하게 설정
    print(contents) #잘 가져와지나 확인
    link = soup.select('body > div.wrapper > div.article > p:nth-child(3) > a') #위와 동일하게 태그가 너무 많아서 상세하게 설정 
    all_name.append(names[0].text) #h1에서 첫번째만 사용
    all_date.append(date.text)
    all_link.append(link)
    all_content.append(contents)
```

```python
print(all_content)
df = pd.DataFrame({'제목' : all_name, '작성날짜': all_date,'링크':all_link,'학습 콘텐츠':all_content}) #데이터프레임 설정
print(df)
```
