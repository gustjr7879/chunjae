```python
import matplotlib.pyplot as plt


import cv2 as cv

imgL = cv.imread('./tsukuba_l.png',0) # 0파라미터를 줘서 2차원으로 불러올 수 있다.
imgR = cv.imread('./tsukuba_R.png',0)
```

```python
plt.imshow(imgL,'gray') #파라미터를 gray로 줘서 흑백으로 표시
```

```python
#스테레오 이미지로 스테레오 비전 처리를 해보자 이때는 opencv에서 제공하는 StereoBM_create 함수를 사용한다.
#이 함수는 numDisparitie, blocksize라는 두가지 파라미터를 입력받는다.
#num- 은 알고리즘에 사용될 disparity 탐색범위를 말한다. 
#blocksize는 가운데 픽셀을 기준으로 탐색하는 블록크기를 결정한다.
```

```python
stereo = cv.StereoBM_create(numDisparities = 32, blockSize = 21)
disparity = stereo.compute(imgL,imgR)

print(disparity) #연산된 값은 array형식으로 된다.
```

```python
plt.imshow(disparity,'gray')
```
