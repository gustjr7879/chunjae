```python
import pandas as pd

dropout = pd.read_csv('dropout_processed.csv',encoding='cp949').drop('Unnamed: 0',axis=1)
```

```python
print(dropout.isna().sum())
```

```python
import matplotlib.pyplot as plt
print(dropout['Target'].value_counts())
plt.pie(dropout['Target'].value_counts(sort=False),labels=dropout['Target'].unique())
plt.show()
```

```python
import plotly.express as px

fig = px.pie(dropout,names='Target')
fig.show()
```

```python
import seaborn as sns
px.bar(dropout, x="Gender",color='Target',barmode='group')
```

```python
sns.countplot(dropout, x="Gender",hue='Target')
```

```python
print(dropout['Marital status'].unique())
married_list = ['Single','Married','Widower','Divorced','Facto Union','Legally Seperated']

dropout['Marital status'] = dropout['Marital status'].apply(lambda x : married_list[int(x)-1])
sns.countplot(dropout,x='Marital status',hue='Target')
```

```python
px.bar(dropout,'Marital status',color='Target',barmode='group')
```

```python
sns.histplot(dropout,y='Course',hue='Target',multiple='stack')
plt.show()

sns.histplot(dropout,y='Course',hue='Target',multiple='fill')
plt.show()
```

```python
sns.histplot(dropout,x='Nationality',hue='Target',multiple='stack')
plt.show()
sns.histplot(dropout,x='Nationality',hue='Target',multiple='fill')
plt.show()
```

```python
sns.countplot(dropout,x = 'Age at enrollment')
```

```python
dfdf = dropout[['Scholarship holder','Unemployment rate','Inflation rate','GDP']]

print(dfdf.dtypes)
dfdf = dfdf.corr()
sns.heatmap(dfdf,annot=True)
```
